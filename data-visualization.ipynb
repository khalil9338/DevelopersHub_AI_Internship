```python
# Task 1: Exploring and Visualizing a Simple Dataset
# DevelopersHub Corporation - AI/ML Internship

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================
# Step 1: Load the dataset
# ============================================
# Using seaborn's built-in tips dataset
df = sns.load_dataset('tips')

# ============================================
# Step 2: Inspect the dataset
# ============================================
print("="*50)
print("DATASET EXPLORATION")
print("="*50)

# Shape of dataset
print(f"\n1. Shape of dataset: {df.shape}")

# Column names
print(f"\n2. Column names: {list(df.columns)}")

# First few rows using head()
print(f"\n3. First 5 rows:")
print(df.head())

# Using info()
print(f"\n4. Dataset Info:")
df.info()

# Using describe()
print(f"\n5. Summary Statistics:")
print(df.describe())

# ============================================
# Step 3: Visualizations
# ============================================

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(15, 12))

# 3.1 Scatter plot - relationship between features
plt.subplot(2, 2, 1)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', size='size')
plt.title('Scatter Plot: Total Bill vs Tip', fontsize=14)
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# 3.2 Histogram 1 - Distribution of total_bill
plt.subplot(2, 2, 2)
plt.hist(df['total_bill'], bins=20, color='steelblue', edgecolor='black')
plt.title('Histogram: Total Bill Distribution', fontsize=14)
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

# 3.3 Histogram 2 - Distribution of tip
plt.subplot(2, 2, 3)
plt.hist(df['tip'], bins=20, color='coral', edgecolor='black')
plt.title('Histogram: Tip Distribution', fontsize=14)
plt.xlabel('Tip ($)')
plt.ylabel('Frequency')

# 3.4 Box plot - identify outliers
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='day', y='total_bill')
plt.title('Box Plot: Total Bill by Day (Outliers visible)', fontsize=14)
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

plt.tight_layout()
plt.show()

# ============================================
# Step 4: Additional Box Plot for tips
# ============================================
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='time', y='tip')
plt.title('Box Plot: Tip by Meal Time', fontsize=14)
plt.xlabel('Meal Time')
plt.ylabel('Tip ($)')
plt.show()

print("\n" + "="*50)
print("TASK 1 COMPLETED SUCCESSFULLY!")
print("="*50)
```
